```python
class RMSProp():
    def __init__(self, parameters, learning_rate, alpha=0.9, eps=1e-8):
        self.parameters = list(parameters)
        self.learning_rate = learning_rate
        self.alpha = alpha
        self.eps = eps

        self.v = [0 for _ in self.parameters]


    def step(self):

        for i,p in enumerate(self.parameters):
            g = p.grad

            self.v[i] = self.alpha * self.v[i] + (1 - self.alpha) * (g**2)

            p.data -= self.learning_rate * (g / ((self.v[i] ** 0.5) + self.eps))


class Adam():
    def __init__(self, parameters, learning_rate, beta1=0.9, beta2=0.999, eps=1e-8):
        self.parameters = list(parameters)
        self.learning_rate = learning_rate
        self.beta1 = beta1
        self.beta2 = beta2
        self.eps = eps

        self.t = 0

        self.m = [0 for _ in self.parameters]
        self.v = [0 for _ in self.parameters]
        

    def step(self):
        self.t += 1

        for i,p in enumerate(self.parameters):
            g = p.grad

            self.m[i] = self.beta1 * self.m[i] + (1 - self.beta1) * g
            self.v[i] = self.beta2 * self.v[i] + (1 - self.beta2) * (g**2)

            m_hat = self.m[i] / (1 - self.beta1 ** self.t)
            v_hat = self.v[i] / (1 - self.beta2 ** self.t)

            p.data -= self.learning_rate * m_hat / ((v_hat ** 0.5) + self.eps)
```
